```python
# Creating bboxes for remaining points that were not resolved with geocoders

import geopandas as gpd
from shapely.geometry import box
from pyproj import Transformer

# Create Bboxes for locations that does not yet have any

# Manually select by id

ids = [1411, 10061]

def create_bbox_ids(lat, lon):
    # WGS84 Has to be converted to projected coords
    transformer = Transformer.from_crs("EPSG:4326", "EPSG:3857", always_xy = True)
    x, y = transformer.transform(lon, lat)

    # BBox specification
    bbox = box(x - 5000, y - 5000, x + 5000, y + 5000)

    # Convert back to WGS84
    transform_back = Transformer.from_crs("EPSG:3857", "EPSG:4326", always_xy = True)
    west, south = transform_back.transform(bbox.bounds[0], bbox.bounds[1])
    east, north = transform_back.transform(bbox.bounds[2], bbox.bounds[3])

    return f"{{'east': {east}, 'south': {south}, 'north': {north}, 'west': {west}}}"

#Apply for selected rows based on id-column

eval_pack.loc[eval_pack['id'].isin(ids), 'bbox'] = eval_pack[eval_pack['id'].isin(ids)].apply(
    lambda row : create_bbox_ids(row['lat'], row['lon']), axis = 1
)
eval_pack
```

```python
# Parsing GN bbox to JSON
# I used ChatGPT4 for help here

def parse_bbox_GN(bbox):
    if isinstance(bbox, dict):
        return {k: float(v) if v is not None else None for k,v in bbox.items()} # THIS USED TO BE EXTENSION: if k != 'accuracyLevel'} # Dropping the accuracylevel infomration

    if isinstance(bbox, str):
        try:
            bbox = bbox.replace("'", '"')
            bbox_dict = json.loads(bbox)
            return {k: float(v) if v is not None else None for k, v in bbox_dict.items()} # THIS USED TO BE EXTENSION: if k != 'accuracyLevel'}
        except (json.JSONDecodeError, ValueError, TypeError):
            return None
            
    return None
```

```python
# Parsin WoF bbox to JSON
# This code was used earlier when the WoF format was structured as [yx, yx] instead of json-like format

def parse_bbox_WoF(bbox):
    if isinstance(bbox, str):
        try:
            bbox = json.loads(bbox)
        except json.JSONDecodeError:
            return None
       
    if isinstance(bbox, list) and len(bbox) == 4: # To standardize the bbox structure
        try:
            return {
                    "west": float(bbox[0]),
                    "south": float(bbox[1]), 
                    "east" : float(bbox[2]), 
                    "north" : float(bbox[3])
            }
        except (ValueError, TypeError):
            return None
    return None
```

```python
# Overlay analysis to calculate points inside BBoxes
def is_point_in_bbox (lat, lon, bbox):
    if not isinstance(bbox, dict) or not bbox:
        return False


    if any(value is None for value in bbox.values()):
        return False

    lat = float(lat)
    lon = float(lon)
    
    return bbox.get("south", float('-inf')) <= lat <= bbox.get("north", float('inf')) and \
            bbox.get("west", float('-inf')) <= lon <= bbox.get("east", float('inf'))
```

```python
# Creating folium map
# BBoxes must be parsed every time imported from csv before folium - it does not recognize separate objects when they are in string

import folium
#from folium.plugins import MarkerCluster

madagascar_center = (-20.0, 47.0)

color_map = {1: 'green', 0: 'red'}

m = folium.Map(location = madagascar_center, zoom_start = 5.45)
#marker_cluster = MarkerCluster().add_to(m)

def connect_points(lat1, lon1, lat2, lon2, group):
    if (lat1,lon1) == (0,0) or (lat2,lon2) == (0,0):
        return
    folium.PolyLine(
        locations = [(lat1, lon1), (lat2, lon2)],
        color = 'gray',
        weight = 2
    ).add_to(group)
        
# BBox:
def add_bbox_to_map(bbox, map_obj):
    if bbox:
        folium.Rectangle(
            bounds = [(bbox['south'], bbox['west']),(bbox['north'], bbox['east'])],
            tooltip=(row['id']),
            color = 'black', weight = 3, opacity = 0.3, fill = False
        ).add_to(map_obj)


# Points:
Pelias = folium.FeatureGroup('Pelias').add_to(m)
Nominatim = folium.FeatureGroup('Nominatim').add_to(m)
GeoNames = folium.FeatureGroup('GeoNames').add_to(m)

for _, row in eval_pack.iterrows():
    folium.CircleMarker(
        location = (row['p_lat'], row['p_lon']),
        radius = 6,
        color = 'orange',
        fill = True,
        fill_color = color_map.get(row['combine_pel']),
        fill_opacity = 0.4,
        popup = f"Name: {row['place_names']} | INSIDE: {row['combine_pel']}"
    ).add_to(Pelias)
    
for _, row in eval_pack.iterrows():
    folium.CircleMarker(
        location = (row['n_lat'], row['n_lon']),
        radius = 6,
        color = 'blue',
        fill = True,
        fill_color = color_map.get(row['combine_nom']),
        fill_opacity = 0.4,
        popup = f"Name: {row['place_names']} | INSIDE: {row['combine_nom']}"
    ).add_to(Nominatim)
    
for _, row in eval_pack.iterrows():
    folium.CircleMarker(
        location = (row['g_lat'], row['g_lon']),
        radius = 6,
        color = 'purple',
        fill = True,
        fill_color = color_map.get(row['combine_gn']),
        fill_opacity = 0.4,
        popup = f"Name: {row['place_names']} | INSIDE: {row['combine_gn']}"
    ).add_to(GeoNames)
    
for _, row in eval_pack.iterrows():
    folium.CircleMarker(
        location = (row['lat'], row['lon']),
        radius = 10,
        color = 'red',
        opacity = 0.6,
        weight = 2.5,
        fill = False,
        popup = f"Name: {row['annot_names']} | ID: {row['id']}"
    ).add_to(m)


for _, row in eval_pack.iterrows():   
    add_bbox_to_map(row['bbox_WoF'], m)
    add_bbox_to_map(row['bbox_GN'], m)
    add_bbox_to_map(row['bbox'], m)

    connect_points(row ['p_lat'], row ['p_lon'], row ['lat'], row ['lon'], Pelias)
    connect_points(row ['n_lat'], row ['n_lon'], row ['lat'], row ['lon'], Nominatim)   
    connect_points(row ['g_lat'], row ['g_lon'], row ['lat'], row ['lon'], GeoNames)  


folium.LayerControl().add_to(m)
```
